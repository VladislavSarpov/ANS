```python
for col, dtype in data.dtypes.items():
    vc = dict(data[col].value_counts(dropna=False))
    if np.nan in vc:
        print(vc[np.nan], '\t', dtype, col, data[col].unique())
```

```python
data['MSZoning'] = data['MSZoning'].fillna('NaN')
data['Alley'] = data['Alley'].fillna('No Alley')
data['Utilities'] = data['Utilities'].fillna('No Utilities')
data['MasVnrType'] = data['MasVnrType'].fillna('None')
data['BsmtQual'] = data['BsmtQual'].fillna('No BsmtQual')
data['BsmtCond'] = data['BsmtCond'].fillna('No BsmtQual')
data['BsmtExposure'] = data['BsmtExposure'].fillna('No BsmtExposure')
data['BsmtFinType1'] = data['BsmtFinType1'].fillna('No BsmtFinType1')
data['BsmtFinType2'] = data['BsmtFinType2'].fillna('No BsmtFinType2')
data['FireplaceQu'] = data['FireplaceQu'].fillna('No FireplaceQu')
data['MiscFeature'] = data['MiscFeature'].fillna('No MiscFeature')
data['GarageType'] = data['GarageType'].fillna('No GarageType')
data['GarageFinish'] = data['GarageFinish'].fillna('No GarageFinish')
data['GarageQual'] = data['GarageQual'].fillna('No GarageQual')
data['GarageCond'] = data['GarageCond'].fillna('No GarageCond')
data['PoolQC'] = data['PoolQC'].fillna('No PoolQC')
data['Fence'] = data['Fence'].fillna('No FenceMiscFeature')

data['SalePrice'] = data['SalePrice'].fillna(0)

data = data.dropna()

data
```

```python
data_train = data[data['SalePrice'] > 0]
data_test = data[data['SalePrice'] == 0]

data_train.shape, data_test.shape
```

```python
from sklearn.preprocessing import LabelEncoder

def get_labeled_df(df, skipna=True):
    res = df.copy()
    for col, dtype in res.dtypes.items():
        if dtype == np.object:
            le = LabelEncoder()
            if skipna:
                ### fit with the desired col, col in position 0 for this example
                fit_by = pd.Series([i for i in res[col].unique() if type(i) == str])
                le.fit(fit_by)
                ### Set transformed col leaving np.NaN as they are
                res[col] = res[col].apply(lambda x: le.transform([x])[0] if type(x) == str else x)
            else:
                res[col][pd.isnull(res[col])]  = 'NaN'
                res[col] = le.fit_transform(res[col])
    return res
```

```python
lbl_train = get_labeled_df(data_train, False).dropna()
lbl_test = get_labeled_df(data_test, False).dropna()
```

```python
X_train = lbl_train.drop('SalePrice', axis=1)
y_train = lbl_train['SalePrice']
X_test = lbl_test
```

```python
from sklearn.ensemble import RandomForestClassifier

clf_rf = RandomForestClassifier(n_estimators=10, max_depth=5, min_samples_leaf=20, max_features=0.5, n_jobs=-1)
clf_rf.fit(X_train, y_train)
clf_rf.score(X_train, y_train)
```

```python
y_pred = clf_rf.predict(X_train)

tmp = lbl_train.copy()
tmp['pred'] = y_pred
tmp['sqr_diff'] = (tmp['pred'] - tmp['SalePrice']) ** 2

tmp
```
